```python
#accessing item
arr = [1,2,3]
print(len(arr))
arr
print(arr[1])
```

```python
#reverse
arr1 = ['A','B','C','D','E']
print(arr1[-1])
```

```python
#Slicing last element -1
arr1 = ['A','B','C','D','E']
print(arr1[1:4])
print(arr1[:4])
print(arr1[2:])
print(arr1[-3:])#using reverse
```

```python
#change item value
arr1 = ['A','B','C','D','E']
arr1[1] = 'F'

# add new item at the end
arr1.append('G')

#insert item
arr1.insert(1,'H')

#extend list
arr1.extend('B')
print(arr1)

#searching item
'A' in arr1
'K' in arr1
```

```python
#repetition
arr2 = ["Aadi"]
arr2 = arr2*6
print(arr2)
```
